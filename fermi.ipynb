```python
import numpy as np
import matplotlib.pyplot as plt
import sobol_seq
```

```python
s = 0.5
g = 2 * s + 1
V = 100.0
N = 10.0
N_p = 1000
d = 3

def N_p_density(p, T, d = 3, mu = 1):
    e = 0
    for i in range(d):
        e += 0.5 * p[i]**2
    return (g * V / (1 + np.exp((e - mu) / T))) / ((2 * np.pi)**d)

def E_density(p, T, d = 3, mu = 1):
    e = 0
    for i in range(d):
        e += 0.5 * p[i]**2
    return (e * N_p_density(p, T, d, mu))

def N_calc(mu, T, d=3):
    a = sobol_seq.i4_sobol_generate(d, N_p)
    scale = 10.0 + 10.0 * T**0.5
    
    for i in range(N_p):
        for j in range(d):
            a[i][j] *= scale
            
    res = 0.0
    for i in range(N_p):
        res += N_p_density(a[i], T, d, mu)
    res /= N_p
    res *= scale**d
    res *= 2.0**d
    return res
```

```python
def mu_T(T, d = 3, err=1e-3):
    l = -1e6
    r = 1e6
    while (r - l > err):
        m = (l + r) / 2.0
        if (N_calc(m, T, d) > N):
            r = m
        else:
            l = m
    return (l + r) / 2.0
```

```python
s = 0.5
g = 2 * s + 1
V = 100.0
N = 10.0
N_p = 1000
d = 3

a = sobol_seq.i4_sobol_generate(d, N_p)

T = np.linspace(-5.0, 3.0, 101)
T = np.exp(T)
E_N = []
mu_val = []
for t in T:
    Ec = 0.0
    mu = mu_T(t, d)
    mu_val.append(mu)
    #print(N_calc(mu, t, d))
    scale = 10.0 + 10.0 * t**0.5
    b = a.copy()
    for i in range(N_p):
        for j in range(d):
            b[i][j] *= scale
    
    for i in range(N_p):
        Ec += E_density(b[i], t, d, mu)
    Ec /= N_p
    Ec *= scale**d
    Ec *= 2.0**d
    Ec /= N
    E_N.append(Ec)
    
plt.plot(T, E_N)
#plt.plot(T, mu_val)
```

```python
plt.plot(T, mu_val)
```
